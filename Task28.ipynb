```python
#маємо вибірку, розподілену за нормальним законом, n=20, y=0.95
n=20
y=0.95
x1=1
x2=2
x3=4
x4=6
x5=7
n1=2
n2=3
n3=10
n4=3
n5=2
#спочатку знайдемо середнє математичне відхилення:
#вибіркове середнє xв буде:
xs=(x1*n1+x2*n2+x3*n3+x4*n4+x5*n5)/n
print("Вибіркове середнє дорівнює", xs)
#тепер дисперсію:
dsp=(((x1-xs)**2)*n1+((x2-xs)**2)*n2+((x3-xs)**2)*n3+((x4-xs)**2)*n4+((x5-xs)**2)*n5)/(n-1)
print("Дисперсія дорівнює", round(dsp, 1))
#тепер середнє квадратичне відхилення:
s=dsp**0.5
print("середнє квадратичне відхилення дорівнює", round(s, 1))
#з таблиці значень при вже даних n=20 і у=0.95 знаходимо q=0.37, далі довірчий інтервал:  s(1-q)<σ<s(1+q):
q=0.37
left=s*(1-q)
right=s*(1+q)
print(round(left,2), "<σ<", round(right,2))




```

```python
#маємо знайти з надійністю у=0.95 інтервальну оцінку ймовірності події А в кожному з n=100 незалежних повторних випробувань, 
#якщо подія відбулась m=40 раз
#Ф(t)=0,95/2=0,475. З таблиці Лапласа знайдемо t=1.96. Далі всі ці значення підставляємо в формулу:
t=1.96
m=40
n=100
p1=(1/(t**2+n))*(m+((t**2)/2)-t*(((m*(n-m))/n+t**2/4)**0.5))
print(round(p1, 2))
p2=(1/(t**2+n))*(m+((t**2)/2)+t*(((m*(n-m))/n+t**2/4)**0.5))
print(round(p2,2))
# інтервальною оцінкою для ймовірності настання події А буде довірчий інтервал (0,31 ; 0,5)
```

```python
#маємо надійність 0.99, точність оцінки 0.6, середнє квадратичне відхилення 3:
#Ф(t)=0,99/2=0,495. З таблиці Лапласа знайдемо t=2.58. Далі всі ці значення підставляємо в формулу:
sg=3
bt=0.6
t=2.58
n=((t**2)*(sg**2))/bt**2
print("Мінімальний об`єм вибірки = ", round(n, 2))

```
